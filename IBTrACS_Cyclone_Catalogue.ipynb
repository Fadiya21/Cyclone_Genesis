```python
def filter_by_date_range(ibtracs, start_year, end_year):
    return ibtracs[
        (ibtracs['BASIN']=='WP') &
        (ibtracs['ISO_TIME'].dt.year >= start_year)&
        (ibtracs['ISO_TIME'].dt.year <= end_year)
    ]

def get_days_for_month(year, month):
    days_in_month=calendar.monthrange(year, month)[1]
    if month == 2 and days_in_month ==29:
        return [f"{i:02d}" for i in range(1, 30)]
    return [f"{i:02d}" for i in range(1, days_in_month + 1)]

ibtracs = pd.read_csv('/Users/fadiya/Documents/cycone/data/downloads/ibtracs.ALL.list.v04r01.csv')

columns = ['SID', 'SEASON', 'BASIN', 'NAME', 'ISO_TIME', 'LAT', 'LON']
ibtracs = ibtracs[columns]

ibtracs['ISO_TIME']=pd.to_datetime(ibtracs['ISO_TIME'], errors='coerce')
ibtracs=ibtracs.dropna(subset=['ISO_TIME', 'LAT', 'LON'])

start_year = 2002
end_year = 2003
ibtracs = filter_by_date_range(ibtracs, start_year, end_year)
print(f"Number of cyclone records after filtering: {len(ibtracs)}")

ibtracs=ibtracs.sort_values(by=['SID', 'ISO_TIME'])
genesis_df = ibtracs.groupby('SID').first().reset_index()

genesis_df = genesis_df.rename(columns={
    'ISO_TIME': 'genesis_time',
    'LAT': 'genesis_lat',
    'LON':'genesis_lon'
})

print(genesis_df[['SID', 'genesis_time', 'genesis_lat', 'genesis_lon']].head(10))

print("Available BASIN values:", ibtracs['BASIN'].unique())
print("Data range in ISO_TIME:", ibtracs['ISO_TIME'].min(), "to", ibtracs['ISO_TIME'].max())
```
